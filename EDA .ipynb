```python
hmap = businessdata[(businessdata['price']=='$')|(businessdata['price']=='$$')|(businessdata['price']=='$$$')]
```

```python
hmap['price'].value_counts()
```

```python
hmap = hmap.head(10)
```

```python
hmap.columns
```

```python
hmap.set_index('name',inplace=True)
```

```python
hmap.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
hmap.columns
```

```python
hmap.shape
```

```python
hmap.drop_duplicates(subset=['name'],inplace=True)
```

```python
hmap.shape
```

```python
hmap.replace('$$',2,inplace=True)
hmap.replace('$$$',3,inplace=True)
```

```python
hmap['price'].value_counts()
```

```python
d = hmap.pivot('name','rating','price')
```

```python
d
```

```python
sns.set_style("dark")
sns.heatmap(d.head(100),cmap="crest")
```
