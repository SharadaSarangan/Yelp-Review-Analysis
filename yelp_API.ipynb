#### 1.2 Create search function

```python
def search(api_key, term, location):
    """Query the Search API by a search term and location.
    Args:
        term (str): The search term passed to the API.
        location (str): The search location passed to the API.
    Returns:
        dict: The JSON response from the request.
    """
    global data
    data = pd.DataFrame([])
    for i in range(0,100000,50):
        
        url_params = {
            'term': term.replace(' ', '+'),
            'location': location.replace(' ', '+'),
            'limit': SEARCH_LIMIT,
            'offset':i
        }
    
        response=request(API_HOST, SEARCH_PATH, api_key, url_params=url_params)
        print(response)
        cols = list(response['businesses'][0].keys())
        
        

        for biz in response['businesses']:
            data = data.append(biz, ignore_index=True)
      


```

#### 2.1 Create search_review function

```python

def search_review(business_id):
    global data_reviews
    data_reviews = pd.DataFrame()
    for i in business_id:
        fullpath='/v3/businesses/'+i+'/reviews'
        response=request(API_HOST,fullpath, API_KEY)
        cols = list(response['reviews'][0].keys())
        
        for re in response['reviews']:
            data_reviews = data_reviews.append(re, ignore_index=True)
            
```

```python
search_review(business_id)
```

#### 2.2 Break down the nested json

```python
user_nest=pd.json_normalize(data_reviews["user"])
user_nest.rename(columns={"id":"user_id","profile_url":"user_profile","image_url":"image_url","name":"user_name"},inplace=True)
#user_nest.drop
```

```python
full_reviews=pd.concat([data_reviews,user_nest],axis=1)
```

#### 2.3 Split url to get business_key, which can map to user data

```python
full_reviews["split_url"]=full_reviews["url"].str.split("?").apply(lambda y: y[0])
```

```python
full_reviews["business_key"]=full_reviews["split_url"].str.split("https://www.yelp.com/biz/").apply(lambda y: y[1])
```
