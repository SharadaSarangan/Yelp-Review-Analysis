```python
import pandas as pd
import numpy as np
import nltk
import re
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('corpus')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('vader_lexicon')
```

```python
data = pd.read_csv('full_reviews.csv')
```

```python
data.head()
```

```python
reviews = pd.DataFrame(data[['text','rating']])
```

```python
reviews['rating'].value_counts()
```

```python
def assign_label(x):
    if(x<3):
        return -1
    elif(x==3):
        return 0
    else:
        return 1
```

```python
reviews['label'] = reviews['rating'].apply(assign_label)
```

```python
reviews.head()
```

```python
reviews['tokens'] = reviews['text'].apply(lambda x: " ".join(x.lower() for x in str(x).split()))
```

```python
from bs4 import BeautifulSoup
reviews['tokens']=reviews['tokens'].apply(lambda x: BeautifulSoup(x).get_text())
reviews['tokens']=reviews['tokens'].apply(lambda x: re.sub(r"http\S+", "", x))
```

```python
reviews.head()
```

```python
reviews['rating'].value_counts()
```

```python
reviews.head()
```

```python
reviews['tokens'] = reviews['tokens'].apply(lambda x: " ".join([re.sub('[^A-Za-z]+','', x) for x in nltk.word_tokenize(x)]))
```

```python
reviews.head()
```

```python
reviews['tokens']=reviews['tokens'].apply(lambda x: re.sub(' +', ' ', x))
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english') 
reviews['tokens'] = reviews['tokens'].apply(lambda x: " ".join([x for x in x.split() if x not in stop]))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
reviews['tokens']=reviews['tokens'].apply(lambda x: " ".join([lemmatizer.lemmatize(w) for w in nltk.word_tokenize(x)]))
```

```python
reviews.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain, ytest = train_test_split(reviews['tokens'], reviews['label'], test_size=0.20, random_state=101)
```

```python
xtrain.head()
```

```python
xtest.shape
```

```python
len(reviews['tokens'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer= TfidfVectorizer()
tf_xtrain = vectorizer.fit_transform(xtrain)
tf_xtest = vectorizer.transform(xtest)
```

```python
np.array(tf_xtrain)
```

```python
vector = pd.DataFrame.sparse.from_spmatrix(tf_xtrain)
```

```python
vector.shape[1]
```

```python
vectorizer.get_feature_names()
```

```python
d = dict()
column = 0
features = list(vectorizer.get_feature_names())
for i in features:
    d[column] = i
    column+=1
```

```python
vector.rename(d,axis=1,inplace=True)
```
